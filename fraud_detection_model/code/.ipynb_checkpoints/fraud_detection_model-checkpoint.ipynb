### Integrate PySpark to work with your data in a distributed manner, even if you’re starting with a small dataset 

```python
from pyspark.sql import SparkSession

#create a spark session

spark = SparkSession.builder\
        .appName("FraudDetection")\
        .getOrCreate()

# Print a confirmation that Spark is running
print("spark session created: ", spark)
```

```python
#load the data set
# Load the dataset with header and infer the schema automatically

df = spark.read.csv("creditcard.csv", header=True, inferSchema=True)

df.show(5)
```

```python
# Print the schema to understand data types
df.printSchema()
```

```python
# Get basic statistics for numerical columns
df.describe().show()
```

```python
#convert to pandas for taking a peek at the dataset
df.limit(100).toPandas()

#above can also be a strategy in real life, may be slice a the data to look at it in pandas
```
